```python
from scipy.fft import fft, ifft
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.pyplot as plt
```

```python
def interpolate(inp, fi):
    i, f = int(fi // 1), fi % 1  # Split floating-point index into whole & fractional parts.
    j = i+1 if f > 0 else i  # Avoid index error.
    return (1-f) * inp[i] + f * inp[j]

def get_mse(real, transformed):
    mse = (np.linalg.norm(real-transformed)**2)/len(real)
    return mse
```

# $\Pi$ signal

```python
t = np.linspace(0, 1, 2000, endpoint=False)
s = signal.square(2 * np.pi * 10 * t)+1

mses = []
for n in range(1,200):
    s_prime = s[0::n]
    t_prime = np.linspace(0, 1, len(s_prime), endpoint=False)
    y = fft(s_prime)
    s_prime = ifft(y).astype(float)
    s_prime = ifft(y).astype(float)
    new_len = len(t)
    delta = (len(s_prime)-1) / (new_len-1)
    s_interpolated = [interpolate(s_prime, i*delta) for i in range(new_len)]
    mses.append(get_mse(s, s_interpolated))
plt.plot(mses)
plt.show()

plt.plot(t, s)
plt.ylim(-1, 3)
plt.show()
```
